```python
# Need MongoDb database called "travisCAD"
# Action REQUIRED!!  Line 27vpyautogui.press("must enter string")

#Dependencies and blank Lists
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as bs
import pyautogui
from datetime import date
import time
import datetime


ptype = []
addy = []
appvalue = []
appyear = []
Nextlink = ''
current_page = 1

executable_path = {'executable_path' : 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
url = 'http://propaccess.traviscad.org/clientdb/?cid=1'
print("Opening Site to Scrape")
browser.visit(url)
time.sleep(6)
browser.find_by_id('propertySearchOptions_searchText').click()
pyautogui.press("7")
#pyautogui.typewrite(x)
time.sleep(2)
browser.find_by_name('propertySearchOptions:search').click()

print("Identifying Number of Pages")
soup = bs(browser.html, 'html.parser')

B = soup.find_all('td', class_="paging")
A = soup.find_all('a', href=True)
tl_pages = int(A[-1].text)+1

def update_table(x):
    #time.sleep(1)
    table = pd.read_html(browser.html)[0]
    df = table.copy()
    #print("Copying Table Data")
    for items in df["Unnamed: 3"]:
        text = str(items)
        if text.find("Page: ") == -1:
            ptype.append(items)
    for items in df["Unnamed: 4"]:
        text = str(items)
        if text.find("Page: ") == -1:
            addy.append(items)
    for items in df["Unnamed: 8"]:
        text = str(items)
        YR_ = date.today().year
        if text.find("Page: ") == -1:
            appvalue.append(items)
            appyear.append(YR_)
    td_time = datetime.datetime.now().time()
    print(f"Table Data Stored: {td_time}")
    return None;

start_time = datetime.datetime.now().time()
print(f"Starting Loop {start_time}")  


for cycles in range(tl_pages):
    try:
        update_table(cycles)
        #print("Locating Next Page Link")
        soup2 = bs(browser.html, 'html.parser')
        B = soup2.find_all('td', class_="paging")
        A = soup2.find_all('a', href=True)
        current_page += 1
        Nextlink = ''
        
        for link in A:
            pg = str(current_page)
            if link.text == pg:
                Nextlink = link['href']

        #pg_time = datetime.datetime.now().time()        
        #print(f"Attempting Link {pg_time}")
        try:
            browser.click_link_by_href(Nextlink)
        except:
            try:
                Nextlink= A[-2]['href']
                browser.click_link_by_href(Nextlink)
            except:
                continue
            brk_time = datetime.datetime.now().time()
            print(f"Link Fail {brk_time}")
            continue
    except:
        break
sc_time = datetime.datetime.now().time()
tt_rr = str(len(appyear))
print(f"Scrub Complete {tt_rr} records at {sc_time}")

```
