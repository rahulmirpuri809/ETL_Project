```python
# Get your zws-id from zillow and put in a config file you can call
# then you Must run this prior
#!pip install pyzillow
#Documentation at https://buildmedia.readthedocs.org/media/pdf/pyzillow/latest/pyzillow.pdf

import pandas as pd
from zillow_config import zwsid
from pyzillow.pyzillow import ZillowWrapper, GetDeepSearchResults

df = pd.read_csv("travisCAD_cleaned.csv")

address = []
zipcodes = []
ApprDist = []
AppType = []


for item in df["Property Address"]:
    address.append(item)
for z in df["Zipcode"]:
    zipcodes.append(z)
for app in df["2019 Appraised Value"]:
    ApprDist.append(app)
for aptype in df["Type"]:
    AppType.append(aptype)

print('Done')
```

```python
print(len(zipcodes))
```

```python
ref = list(zip(address, zipcodes))

zillow_id = []
home_type = []
tax_year = []
tax_value = []
year_built = []
property_size = []
home_size = []
bathrooms = []
bedrooms = []
last_sold_date = []
last_sold_price = []
zestimate_amount = []
zestimate_last_update = []
lat = []
lon = []
hlink = [] 

record = 0
for x,y in ref:
    addy = x
    zcode = y
    record += 1
    try:
        zillow_data = ZillowWrapper(zwsid)
        deep_search_response = zillow_data.get_deep_search_results(addy,zcode)
        result = GetDeepSearchResults(deep_search_response)
        
        R1= result.zillow_id
        R2 = result.home_type
        R3 = result.tax_year
        R4 = result.tax_value
        R5 = result.year_built
        R6 = result.property_size
        R7 = result.home_size
        R8 = result.bathrooms
        R9 = result.bedrooms
        R10 = result.last_sold_date
        R11 = result.last_sold_price
        R12 = result.zestimate_amount
        R13 = result.zestimate_last_updated
        R14 = result.latitude
        R15 = result.longitude
        R16 = result.home_detail_link
     
        zillow_id.append(R1)
        home_type.append(R2)
        tax_year.append(R3)
        tax_value.append(R4)
        year_built.append(R5)
        property_size.append(R6)
        home_size.append(R7)
        bathrooms.append(R8)
        bedrooms.append(R9)
        last_sold_date.append(R10)
        last_sold_price.append(R11)
        zestimate_amount.append(R12)
        zestimate_last_update.append(R13)
        lat.append(R14)
        lon.append(R15)
        hlink.append(R16)
        
        print(str(record)
              
    except:
        zillow_id.append(None)
        home_type.append(None)
        tax_year.append(None)
        tax_value.append(None)
        year_built.append(None)
        property_size.append(None)
        home_size.append(None)
        bathrooms.append(None)
        bedrooms.append(None)
        last_sold_date.append(None)
        last_sold_price.append(None)
        zestimate_amount.append(None)
        zestimate_last_update.append(None)
        lat.append(None)
        lon.append(None)
        hlink.append(None)
        
        print("no Response")
    
#This is going to take some time
print(len(zillow_id))
```

```python
z_df = pd.DataFrame ({"zillow_id" : zillow_id,
                      "Property_address" : address,
                      "Zipcode" : zipcodes,
                      "CAD_Type" : AppType,
                      "home_type" : home_type,
                      "tax_year" : tax_year,
                      "tax_value" : tax_value,
                      "year_built" : year_built,
                      "property_size" : property_size,
                      "home_size" : home_size,
                      "bathrooms" : bathrooms,
                      "bedrooms" : bedrooms,
                      "last_sold_date" : last_sold_date,
                      "last_sold_price" : last_sold_price,
                      "zestimate_last_update" : zestimate_last_update,
                      "zestimate_amount" : zestimate_amount,
                      "CAD_current_Appraisal" : ApprDist,
                      "Latitude" : lat,
                      "Longitude" : lon,
                      "ZillowLink" : hlink
                      })

#If you want to Remove the ones missing zillow Data
#killNaNs = combined_df[ z_df['zillow_id'] == None ].index
#combined_df.drop(killNaNs , inplace=True)



#Get County,City info from Zipcode
df2 = pd.read_csv("zip_translation.csv")

#Merge DataSets and export as CSV
#final_df = pd.merge(z_df, df2, on='Zipcode', how='left')
z_df.to_csv("zoutput1_TxTravisCAD.csv" , index=False, header=True)
z_df.head(3)
```
